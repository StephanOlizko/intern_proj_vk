```python
#Прологарифмируем шкалу и построим график
np.log(df['query_id'].value_counts() + 1).plot(kind='hist', bins=100)
```

```python
#Получили график, похожий на нормальное распределение
#Уберем строки которые лежат за пределами 1.5 межквартильного размаха (IQR) по логарифмированной шкале количества query_id
# Calculate the counts for each query_id
query_counts = df['query_id'].value_counts()

# Log transform the counts and add 1
log_query_counts = np.log(query_counts + 1)

# Calculate the IQR
Q1 = log_query_counts.quantile(0.25)
Q3 = log_query_counts.quantile(0.75)
IQR = Q3 - Q1

# Create a mask for values within the IQR
mask = (log_query_counts > (Q1 - 1.5 * IQR)) & (log_query_counts < (Q3 + 1.5 * IQR))

# Apply the mask to the counts
filtered_query_counts = query_counts[mask]

# Filter the original DataFrame
df1 = df[df['query_id'].isin(filtered_query_counts.index)]

# Plot the value counts
np.log(df1['query_id'].value_counts() + 1).plot(kind='hist', bins=100)
```

```python
#Построим heatmap для корреляции
import seaborn as sns

sns.heatmap(df2.corr())
```

```python
# Вычисляем матрицу корреляции
corr_matrix = df2.corr().abs()

# Находим пары столбцов с высокой корреляцией c другими столбцами (кроме диагонали)
high_corr_var=np.where(corr_matrix>0.8)

# Выводим пары столбцов
high_corr_var=[(corr_matrix.columns[x],corr_matrix.columns[y]) for x,y in zip(*high_corr_var) if x!=y and x<y]

# Удалим столбцы с высокой корреляцией (оставим один из пары)
df3 = df2.drop(columns=[pair[1] for pair in high_corr_var])

df3
```

________

```python
# Пробуем обучить бейзлайн (модель на каждый query_id)
# Построим модель которая будет предсказывать rank внутри одной сессии (query_id)
from sklearn.model_selection import train_test_split

# Используем линейную регрессию
from sklearn.linear_model import LinearRegression

from sklearn.metrics import ndcg_score

# Создаем словарь для хранения моделей для каждого query_id
models = {}

# Создаем списки для хранения предсказаний и истинных значений
scores = []

# Обучаем отдельную модель для каждого query_id
for query_id in df3['query_id'].unique():
    # Выбираем данные для этого query_id
    df_query = df3[df3['query_id'] == query_id]

    # Разделяем данные на обучающую и тестовую выборки
    X_train, X_test, y_train, y_test = train_test_split(df_query.drop(columns=['rank', 'query_id']), df_query['rank'], test_size=0.2, random_state=42)

    # Обучаем модель
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Сохраняем модель в словаре
    models[query_id] = model

    # Предсказываем ранги для тестовой выборки
    y_pred = model.predict(X_test)

    # Cчитаем NDCG
    scores.append(ndcg_score(np.array([y_test]), np.array([y_pred]), k=5))

# Считаем среднее по метриккам 
print(f"NDCG: {np.mean(scores)}")
```

```python
#Попробуем подход с косинусной близостью
from sklearn.metrics.pairwise import cosine_similarity

# Разделяем данные на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(df3.drop(columns=['rank']), df3['rank'], test_size=0.2, random_state=41)

scores = []

for query_id in df3['query_id'].unique():
    # Выбираем данные для этого query_id
    df_query = df3[df3['query_id'] == query_id]

    # Выбираем данные для этого query_id из обучающей выборки
    X_query_train = X_train[X_train['query_id'] == query_id]
    y_query_train = y_train[X_train['query_id'] == query_id]

    # Выбираем данные для этого query_id из тестовой выборки
    X_query = X_test[X_test['query_id'] == query_id]
    y_query = y_test[X_test['query_id'] == query_id]

    # Считаем предскозание для объкта как взвешенная сумма рангов объектов из обучающей выборки с весами, равными косинусной близости
    # объекта из тестовой выборки с объектами из обучающей выборки
    y_pred = []

    for i in range(len(X_query)):
        similarities = cosine_similarity(X_query.iloc[i].values.reshape(1, -1), X_query_train)[0]
        y_pred.append(np.sum(y_query_train * similarities) / np.sum(similarities))

    # Считаем NDCG
    scores.append(ndcg_score(np.array([y_query]), np.array([y_pred]), k=5))

# Считаем среднее по метриккам
print(f"NDCG: {np.mean(scores)}")

```

```python
# Построим модель которая будет предсказывать rank внутри одной сессии (query_id)
from sklearn.model_selection import train_test_split

# Используем градиентный бустинг
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.metrics import ndcg_score

# Создаем словарь для хранения моделей для каждого query_id
models = {}

# Создаем списки для хранения предсказаний и истинных значений
scores = []

# Обучаем отдельную модель для каждого query_id
for query_id in df3['query_id'].unique():
    # Выбираем данные для этого query_id
    df_query = df3[df3['query_id'] == query_id]

    # Разделяем данные на обучающую и тестовую выборки
    X_train, X_test, y_train, y_test = train_test_split(df_query.drop(columns=['rank', 'query_id']), df_query['rank'], test_size=0.2, random_state=42)

    # Обучаем модель
    model = GradientBoostingRegressor()
    model.fit(X_train, y_train)

    # Сохраняем модель в словаре
    models[query_id] = model

    # Предсказываем ранги для тестовой выборки
    y_pred = model.predict(X_test)

    # Cчитаем NDCG
    scores.append(ndcg_score(np.array([y_test]), np.array([y_pred]), k=5))

# Считаем среднее по метриккам 
print(f"NDCG: {np.mean(scores)}")
```

```python
# Pointwise подход к ранжированию

# Попробуем обучить модель на всех данных
# Разделяем данные на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(df3.drop(columns=['rank']), df3['rank'], test_size=0.2, random_state=42)

# Обучаем модель
model = LinearRegression()
model.fit(X_train, y_train)

# Предсказываем ранги для тестовой выборки
# Посчитаем NDCG для каждой сессии
scores = []

for query_id in df3['query_id'].unique():
    # Выбираем данные для этого query_id
    df_query = df3[df3['query_id'] == query_id]

    # Выбираем данные для этого query_id из тестовой выборки
    X_query = X_test[X_test['query_id'] == query_id]
    y_query = y_test[X_test['query_id'] == query_id]

    # Предсказываем ранги
    y_pred = model.predict(X_query)

    if len(y_pred) == 1:
        continue

    # Считаем NDCG
    scores.append(ndcg_score(np.array([y_query]), np.array([y_pred]), k=5))

# Считаем среднее по метриккам
print(f"NDCG: {np.mean(scores)}")
```
